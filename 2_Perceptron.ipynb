# Deep Learning from Scratch - 2.Perceptron

## Implementation

### Simple perceptron

```python
# Example of AND gate
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1 * w1 + x2 * w2
  if tmp <= theta:
    return 0
  else:
    return 1

print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

### Weight and Bias
image.png

```python
import numpy as np
x = np.array([0, 1]) # Input
w = np.array([0.5, 0.5]) # Weight
b = -0.7 # Bias

print(w * x)
print(np.sum(w * x))
print(np.sum(w * x) + b)
```

```python
# Example of AND gate with weight and bias
def AND_WB(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w * x) + b
  if tmp <= 0:
    return 0
  else:
    return 1

# Example of NAND gate with weight and bias
def NAND_WB(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5]) # Only w and b are different with AND
  b = 0.7
  tmp = np.sum(w * x) + b
  if tmp <= 0:
    return 0
  else:
    return 1

# Example of OR gate with weight and bias
def OR_WB(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5]) # Only w and b are different with AND
  b = -0.2
  tmp = np.sum(w * x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

## Limitation: XOR Gate
image.png

Single perceptron cannot handle the XOR problem
* Each group of points of '0' and '1' cannot be divided with one straight line

### Multi-Layer Perceptron
Which can implement the XOR gate...

image.png
image.png

```python
def XOR(x1, x2):
  s1 = NAND_WB(x1, x2)
  s2 = OR_WB(x1, x2)
  y = AND_WB(s1, s2)
  return y

print(XOR(0, 0))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(1, 1))
```

## Python Interpreter

### List

```python
a = [1, 2, 3, 4, 5]
print(a)
print(len(a)) # Length
a[4] = 99 # Substitute component
print(a)
print("")

# Data Slicing
print(a[0]) # First component
print(a[0:2]) # From first to third components
print(a[1:]) # From second to end components
print(a[:3]) # From first to fourth components
print(a[:-2]) # From first to the last to second components
```

### Dictionary

```python
me = {'height': 180}
print(me['height']) # Approach to the component
print("")

me['height'] = 70
print(me)
```

### Bool

```python
hungry = True
sleepy = False
print(type(hungry))
print(not hungry)
print(hungry and sleepy)
print(hungry or sleepy)
```

### Class
Defining class makes enable to make..
* user-set data types
* exclusive functions (methods) and their features

```python
class Human:
  """
  __init__
  클래스를 초기화하는 방법을 정의함(생성자, constructor라고도 함)
  클래스 인스턴스가 만들어질 때 한 번만 사용
  메소드의 첫 번째 인수로 자신(자신의 인스턴스)을 나타내는 self를 명시적으로 쓰는 것이 특징
  """
  def __init__(self, name):
    self.name = name
    print("Initialized!")
  
  def hello(self):
    print("Hello " + self.name + "!")

  def goodbye(self):
    print("Good-bye " + self.name + "!")


"""
Human의 생성자(초기화 메소드)는 name이라는 인수를 받고,
그 인수로 인스턴스 변수인 self.name을 초기화
인스턴스 변수: 인스턴스별로 저장되는 변수로, self 다음에 속성 이름을 써서 작성하거나 접근할 수 있음
"""
h = Human("Sehyeong")
h.hello()
h.goodbye()
```

## Numpy

### Basic mathematics

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])
print(x)
print(x / 2.0)
print(type(x))
print("")

y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

### N-th array

```python
A = np.array([[1, 2], [3, 4]])
print(A)
print(A * 10)
print(A.shape)
print(A.dtype)
print("")

B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
```

### Broadcast

### Approach to Component

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0])
print(X[0][1])
print("")

for row in X:
  print(row)
print("")

X = X.flatten()
print(X)
print("")

print(X[np.array([0, 2, 4])])
print("")

print(X > 15)
print(X[X>15])
```

## Matplotlib

### Simple graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1) # (start, end, interval)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sine & cosine")
plt.legend()
plt.show()
```
